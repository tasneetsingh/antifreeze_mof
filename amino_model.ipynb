```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from keras.optimizers import Nadam
```

```python
def create_model(input_shape):
    inputs = Input(shape=input_shape)
    x = Dense(64, activation='relu')(inputs)
    x = Dense(32, activation='relu')(x)
    outputs = Dense(1)(x)
    model = Model(inputs, outputs)
    model.compile(optimizer=Nadam(), loss='mse')
    return model
```

```python
y = df['Exp. % MGS'].values.flatten()
X= std_amino.values
input_shape = (X.shape[1],)
model_std = create_model(input_shape)
model_std.fit(X, y, epochs=200, batch_size=10, verbose=1)
std_pred = model_std.predict(X)
```

```python
from scipy.stats import pearsonr
mse = np.mean((std_pred.flatten() - y)**2)
pcc, _ = pearsonr(y, std_pred.flatten())
mse,pcc
```
