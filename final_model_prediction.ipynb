```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import rdkit
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
from rdkit.Chem import GraphDescriptors
from rdkit.Chem import Crippen
from rdkit.Chem import Descriptors
from rdkit.Chem import Lipinski
from rdkit.Chem import rdMolDescriptors
```

```python
UiO_66 = "O=C(O)c1ccc(C(=O)O)cc1"
UiO_cooh="O=C(O)c1cc(C(=O)O)c(C(=O)O)cc1C(=O)O"
UiO_nh2="Nc1cc(C(=O)O)ccc1C(=O)O"
UiO_67="O=C(O)c2ccc(c1ccc(C(=O)O)cc1)cc2"
mof="O=C(O)c1cc(C(=O)O)cc(C(=O)O)c1"
oh="O=C(O)c1cc(O)c(C(=O)O)cc1O"
gly="OCC(O)CO"
sul="CS(C)=O"
Uio_66_2="c1cc(ccc1C(=O)[O])C(=O)[O]"
```

```python
def complete(a):
  mol=Chem.MolFromSmiles(a)
  if mol is None:
        return None  # Handle invalid SMILES strings

  if mol.GetNumConformers() == 0:
        AllChem.EmbedMolecule(mol)
        # Attempt to generate a conformer if none exist
        if mol.GetNumConformers() == 0:
            return None  # Still no conformers, handle the issue

  try:
        AllChem.UFFOptimizeMolecule(mol)
  except ValueError:
        return None  # Handle optimization errors

  a=GraphDescriptors.BalabanJ(mol)
  b=GraphDescriptors.BertzCT(mol)
  c=GraphDescriptors.HallKierAlpha(mol)
  d=Crippen.MolLogP(mol)
  e=Descriptors.ExactMolWt(mol)
  f=Descriptors.FpDensityMorgan1(mol)
  g=Descriptors.MaxPartialCharge(mol)
  h=Lipinski.FractionCSP3(mol)
  i=Lipinski.NHOHCount(mol)
  j=Lipinski.NOCount(mol)
  k=rdMolDescriptors.CalcAsphericity(mol)
  l=rdMolDescriptors.CalcChi0n(mol)
  m=rdMolDescriptors.CalcChi1n(mol)
  n=rdMolDescriptors.CalcChi1v(mol)
  o=rdMolDescriptors.CalcEccentricity(mol)
  p=rdMolDescriptors.CalcFractionCSP3(mol)
  q=rdMolDescriptors.CalcHallKierAlpha(mol)
  r=rdMolDescriptors.CalcInertialShapeFactor(mol)
  s=rdMolDescriptors.CalcKappa1(mol)
  t=rdMolDescriptors.CalcKappa2(mol)
  u=rdMolDescriptors.CalcKappa3(mol)
  v=rdMolDescriptors.CalcNPR1(mol)
  w=rdMolDescriptors.CalcNPR2(mol)
  x=rdMolDescriptors.CalcNumAliphaticCarbocycles(mol)
  y=rdMolDescriptors.CalcNumAliphaticHeterocycles(mol)
  z=rdMolDescriptors.CalcNumAliphaticRings(mol)
  aa=rdMolDescriptors.CalcNumAmideBonds(mol)
  bb=rdMolDescriptors.CalcNumAromaticCarbocycles(mol)
  cc=rdMolDescriptors.CalcNumAromaticHeterocycles(mol)
  dd=rdMolDescriptors.CalcNumAromaticRings(mol)
  ee=rdMolDescriptors.CalcNumHBA(mol)
  ff=rdMolDescriptors.CalcNumHBD(mol)
  gg=rdMolDescriptors.CalcNumHeteroatoms(mol)
  hh=rdMolDescriptors.CalcNumHeterocycles(mol)
  ii=rdMolDescriptors.CalcNumLipinskiHBA(mol)
  jj=rdMolDescriptors.CalcNumLipinskiHBD(mol)
  kk=rdMolDescriptors.CalcNumRings(mol)
  ll=rdMolDescriptors.CalcNumRotatableBonds(mol)
  mm=rdMolDescriptors.CalcNumSaturatedCarbocycles(mol)
  nn=rdMolDescriptors.CalcNumSaturatedHeterocycles(mol)
  oo=rdMolDescriptors.CalcNumSaturatedRings(mol)
  pp=rdMolDescriptors.CalcPBF(mol)
  qq=rdMolDescriptors.CalcPMI1(mol)
  rr=rdMolDescriptors.CalcSpherocityIndex(mol)
  ss=rdMolDescriptors.CalcTPSA(mol)

  value=[a,b,c,d,e,f,g,h,i,j,k,l,m,n,o,p,q,r,s,t,u,v,w,x,y,z,aa,bb,cc,dd,ee,ff,gg,hh,ii,jj,kk,ll,mm,nn,oo,pp,qq,rr,ss]
  return value
```

```python
a=complete(UiO_66)
b=complete(UiO_cooh)
c=complete(UiO_nh2)
d=complete(UiO_67)
e=complete(mof)
f=complete(oh)
g=complete(gly)
h=complete(sul)
i=complete(Uio_66_2)
x=[a,b,c,d,e,f,g,h,i]
df=pd.DataFrame(x)
```

```python
import tensorflow
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from keras.optimizers import Adagrad,Nadam,Adam,RMSprop
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
```

```python
def create_model(input_shape):
    inputs = Input(shape=input_shape)
    x = Dense(100, activation='relu')(inputs)
    x = Dense(100, activation='relu')(x)
    x = Dense(100, activation='relu')(x)
    outputs = Dense(1)(x)
    model = Model(inputs, outputs)
    model.compile(optimizer=Nadam(learning_rate=0.005), loss='mse',)
    return model
```
